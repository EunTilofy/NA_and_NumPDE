```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2.17572 + 2.67041*x - 0.238444*x**2

xs = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
ys = [2.9,2.7,4.8,5.3,7.1,7.6,7.7,7.6,9.4,9.0,9.6,10.0,10.2,9.7,8.3,8.4,9.0,8.3,6.6,6.7,4.1]

plt.scatter(xs, ys, color='blue', label='Data Points')
x = np.linspace(0, 10, 1000)
y = [f(t) for t in x]
plt.plot(x, y, color='red', label='Fitted Curve')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np

G = np.array([[21, 105, 717.5],
              [105, 717.5, 5512.5],
              [717.5, 5512.5, 45166.6]])

singular_values = np.linalg.svd(G, compute_uv=False)

condition_number = np.max(singular_values) / np.min(singular_values)

print("Condition number of G:", condition_number)
```
