```python
import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_csv("Output_A.csv")

data = pd.DataFrame(file)
plt.figure(figsize=(10, 7))
plt.plot(data['x'], data['f'], label='f(x)')
plt.plot(data['x'], data['g'], label='g(x)')
plt.plot(data['x'], data['h'], label='h(x)')

plt.legend()
plt.show()
```
